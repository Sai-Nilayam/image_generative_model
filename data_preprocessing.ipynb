## Making the Feature Label Pairs 

```python
# In our case we need to design three of them, X_1, X_2, Y
# X_1 = Feet, X_2 = Sleeper, Y = Feet with Sipper

# In this step we are going to find all the combinations that are valid.
# Feet, Sleeper, Feet with Sleeper

# Grabing all the names from the source folder
# Lists all the file names from the Source folder
source_path = 'data/training/all_data_named_resized'

files = os.listdir(source_path)
files = [file for file in files if file[-1] != '~']

print(len(files))

# Sorting all the file names in files array.
files.sort()
```

```python
feets = [ele for ele in files if ele[0:2] == 'f_']
# print(feets)
print(len(feets))

sleepers = [ele for ele in files if ele[0:2] == 's_']
# print(sleepers)
print(len(sleepers))

feet_sleepers = [ele for ele in files if ele[0:2] == 'sf']
# print(feet_sleepers)
print(len(feet_sleepers))
```

```python
pairs = []

# Making the pairs
for feet in feets:
    for sleeper in sleepers:
        for foot_sleeper in foot_sleepers:
            # Applying the condition to be a valid pair
            parameter_1 = feet.split('_')[1]
            parameter_2 = sleeper.split('_')[1]
            parameter_3 = foot_sleeper.split('_')[1]
            parameter_4 = foot_sleeper.split('_')[2]
            if (parameter_1==parameter_3) and (parameter_2==parameter_4):
                inst_arr = []
                inst_arr.append(feet)
                inst_arr.append(sleeper)
                inst_arr.append(foot_sleeper)
                pairs.append(inst_arr)
                
print(pairs[0])
print(len(pairs))
```

```python
# Checking all the pairs are correct or not
index = 0
pair = pairs[index]

foot = pair[0]
sleeper = pair[1]
foot_sleeper = pair[2]

source_path = 'data/training/all_data_named_resized/'

foot_img = cv2.imread(source_path + foot)
foot_img = cv2.cvtColor(foot_img, cv2.COLOR_BGR2RGB)

sleeper_img = cv2.imread(source_path + sleeper)
sleeper_img = cv2.cvtColor(sleeper_img, cv2.COLOR_BGR2RGB)

foot_sleeper_img = cv2.imread(source_path + foot_sleeper)
foot_sleeper_img = cv2.cvtColor(foot_sleeper_img, cv2.COLOR_BGR2RGB)

plt.imshow(foot_img)
plt.show()

plt.imshow(sleeper_img)
plt.show()

plt.imshow(foot_sleeper_img)
plt.show()

# The pairs are correct.
```
