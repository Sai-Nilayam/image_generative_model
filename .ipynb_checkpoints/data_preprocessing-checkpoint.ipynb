```python
import os
import matplotlib.pyplot as plt
import cv2
```

```python
# Lists all the file names from the Source folder
base_path = 'data/training/all_data'
# Defining the destination folder.
destination_path = 'data/training/inst_data'

files = os.listdir(base_path)
files = [file for file in files if file[-1] != '~']

print(len(files))

# No. of files = 2*8 + 10*6
print(2*8 + 10*6)

# Sorting all the file names in files array.
files.sort()
```

```python
# Vizualizing each files
index = 16
file_name = files[index]
file_path = base_path + '/' + file_name
img = cv2.imread(file_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```python
# The next set of Images would be names as f_0, fs_0_0 
# Now renaming the person
for i in range(16, len(files), 6):
    
    inst_file_name = files[i]
    inst_file_path = base_path + '/' + inst_file_name
    inst_img = cv2.imread(inst_file_path)
    
    inst_new_file_path = destination_path + '/' + 's_1_{}'.format(i-8) + '.jpg'
    
    cv2.imwrite(inst_new_file_path, inst_img)
    
```
